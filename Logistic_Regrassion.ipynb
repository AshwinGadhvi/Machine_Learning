```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

from sklearn.model_selection import GridSearchCV
```

```python
# Step 2: Load the dataset
file_path = 'SocialNetworkAds.csv'
df = pd.read_csv(file_path)
print(df)
```

```python
# Step 3: Dependent and independent differ
# Assuming the dependent variable is the last column and the rest are independent variables
X = df.iloc[:, :-1].values  # Independent variables
y = df.iloc[:, -1].values   # Dependent variable
```

```python
# Step 4: Reshape if needed
# This step might be required for specific datasets, but let's assume no reshape is needed for now
```

```python
# Step 5: Check shapes
print(X.shape)  # Should print (400, n) where n is the number of features
print(y.shape)  # Should print (400,)
```

```python
# Step 6: Split dataset (sklearn.model_selection)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Step 7: Initialize the model (Logistic Regression)
model = LogisticRegression()
```

```python
# Step 8: Fit the model on the training data
model.fit(X_train, y_train)
```

```python
# Step 9: Predict using the model on the testing data
y_pred = model.predict(X_test)
```

```python
# Step 10: Confusion matrix (original, predicted)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# Step 11: Classification report (original, predicted)
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

```python
tuned_parameters = {
    'C':[10**-4,10**2,10**0,10**2,10**4],
    'penalty':['L2','L1'],
    'solver':['liblinear','lbfgs'],
    'max_iter':[100,200,300]
}
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=.8)
```

```python
model = GridSearchCV(LogisticRegression(),tuned_parameters,cv=5)
```

```python
model.fit(x_train,y_train)
print(model.best_estimator_)
print(model.score(x_test,y_test))
```
